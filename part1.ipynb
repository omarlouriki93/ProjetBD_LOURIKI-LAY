```python
import pandas as pd
tab = pd.read_csv("consommation-electrique-par-secteur-dactivite-commune.csv", low_memory=False, sep=";")
display(tab)
```

```python
tab.describe()
```

```python
conso_df = tab[["Nom région","Conso moyenne Résidentiel (MWh)"]]
display(conso_df)
```

```python
#Calcul de la consommation moyenne résidentiel par région

conso_perComune = conso_df.groupby(['Nom région']).mean()
display(conso_perComune)
```

```python
%matplotlib inline

conso_perComune.plot(kind="bar", title='Consommation moyenne résidentiel par région')
```

```python
#Calcul de la consommation moyenne professionnel par région

tab_NonResidential = tab[["Nom région","Conso moyenne Professionnel (MWh)"]]
conso_perComune_NonResidential = tab_NonResidential.groupby(['Nom région']).mean()
display(conso_perComune_NonResidential)
```

```python
%matplotlib inline

conso_perComune_NonResidential.plot(kind="bar", color="green", title='Consommation moyenne professionnel par région')
```

```python
#Calcul du nombre d'occurence de chaque région dans la table à l'aide de bokeh pour évaluer la précision des données

import pandas
from bokeh.plotting import figure
from bokeh.io import show

mydata = tab['Nom région']
count = mydata.value_counts()

print(count.keys())

plot=figure(x_range=list(count.keys().values), title='Density_of residence and professional sites by region', x_axis_label = "Région", y_axis_label = "Nombre d'enregistrements", background_fill_color="#E8DDCB")
plot.vbar(x=list(count.keys().values), width=0.5, bottom=0, top=count.get_values())

show(plot)

```

```python
#Calcul du nombre de sites résidentiel par région

tab_nbSResi_parRegion_df = tab[["Nom région","Nb sites Résidentiel"]].groupby(['Nom région']).sum()
display(tab_nbSResi_parRegion_df)
```

```python
#Calcul du nombre de sites professionnel par région

tab_nbSPro_parRegion_df = tab[["Nom région","Nb sites Professionnel"]].groupby(['Nom région']).sum()
display(tab_nbSPro_parRegion_df)
```

```python
#affichage des camemberts pour chaques dataframes.

%matplotlib inline

tab_nbSResi_parRegion_df.plot(kind="pie", title="Nombre total de sites résidentiel par région de 2011 à 2016", subplots=True, legend=False)

tab_nbSPro_parRegion_df.plot(kind="pie", title="Nombre total de sites professionnel par région de 2011 à 2016", subplots=True, legend=False)
```

```python
#Calcul de l'évolution du nombre de sites résidentiel

tab_nbSRes_parAnnee_df = tab[["Année","Nb sites Résidentiel"]].groupby(['Année']).mean()
display(tab_nbSRes_parAnnee_df)
```

```python
#Calcul de l'évolution du nombre de sites professionnel

tab_nbSPro_parAnnee_df = tab[["Année","Nb sites Professionnel"]].groupby(['Année']).mean()
display(tab_nbSPro_parAnnee_df)
```

```python
#Modélisation des données à l'aide d'une courbe

%matplotlib inline

tab_nbSRes_parAnnee_df.plot(kind="line", title="Evolution de la moyenne du nombre de sites résidentiel de 2011 à 2016", subplots=True, legend=True)
```

```python
%matplotlib inline

tab_nbSPro_parAnnee_df.plot(kind="line", title="Evolution de la moyenne du nombre de sites professionnel de 2011 à 2016", subplots=True, legend=True, color="red")
```

```python
#Calcul du taux de chauffage électrique moyen par région

tab_CEMoyen_parRegion_df = tab[["Nom région","Taux de chauffage électrique"]].groupby(['Nom région']).mean()
display(tab_CEMoyen_parRegion_df)
```

```python
%matplotlib inline

tab_CEMoyen_parRegion_df.plot(kind="bar", color="orange", title='Taux de chauffage électrique moyen par région')
```

```python
#Calcul de l'évolution de la consommation totale résidentielle à La Chapelle-Saint-Maurice

evolutionConsoTR_LaChapelleSaintMaurice = tab.loc[tab['Nom commune']=="La Chapelle-Saint-Maurice"]
tab_evolution_by_year_Res = evolutionConsoTR_LaChapelleSaintMaurice[["Année","Conso totale Résidentiel (MWh)"]].groupby(['Année']).mean()
display(tab_evolution_by_year_Res)

tab_evolution_by_year_Res.plot(kind="line", title="Evolution de la consommation totale résidentiel à La Chapelle-Saint-Maurice de 2011 à 2016", subplots=True, legend=True, color="purple")
```

```python
#Calcul de l'évolution de la consommation totale professionnelles à La Chapelle-Saint-Maurice

%matplotlib inline

evolutionConsoTP_LaChapelleSaintMaurice = tab.loc[tab['Nom commune']=="La Chapelle-Saint-Maurice"]
tab_evolution_by_year_Pro = evolutionConsoTP_LaChapelleSaintMaurice[["Année","Conso totale Professionnel (MWh)"]].groupby(['Année']).mean()
display(tab_evolution_by_year_Pro)

tab_evolution_by_year_Pro.plot(kind="line", title="Evolution de la consommation totale professionnel à La Chapelle-Saint-Maurice de 2011 à 2016", subplots=True, legend=True, color="green")
```

```python
#Calcul de l'évolution du nombre de sites professionnel à La Chapelle-Saint-Maurice
#On remarque que la courbe ne justifie pas l'évolution drastique de la consommation totale professionnel à La Chapelle-Saint-Maurice

%matplotlib inline

evolutionNbSP_LaChapelleSaintMaurice = tab.loc[tab['Nom commune']=="La Chapelle-Saint-Maurice"]
tab_evolution_by_year_Pro = evolutionNbSP_LaChapelleSaintMaurice[["Année","Nb sites Professionnel"]].groupby(['Année']).mean()
display(tab_evolution_by_year_Pro)

tab_evolution_by_year_Pro.plot(kind="line", title="Evolution de la consommation totale professionnel à La Chapelle-Saint-Maurice de 2011 à 2016", subplots=True, legend=True, color="green")
```

```python
#Calcul de l'évolution de la consommation moyenne professionnelle à La Chapelle-Saint-Maurice
#On déduit que l'évolution drastique de la consommation totale professionnel à La Chapelle-Saint-Maurice est dû à l'augmentation de la consommation moyenne professionnelle

%matplotlib inline

evolutionConsoMP_LaChapelleSaintMaurice = tab.loc[tab['Nom commune']=="La Chapelle-Saint-Maurice"]
tab_evolution_by_year_Pro = evolutionConsoMP_LaChapelleSaintMaurice[["Année","Conso moyenne Professionnel (MWh)"]].groupby(['Année']).mean()
display(tab_evolution_by_year_Pro)

tab_evolution_by_year_Pro.plot(kind="line", title="Evolution de la consommation moyenne professionnel à La Chapelle-Saint-Maurice de 2011 à 2016", subplots=True, legend=True, color="green")
```

```python
#Calcul du nombre des différents types de résidences principales à La Chapelle-Saint-Maurice
#On remarque que la grande majorité des résidences principales datent d'entre 1919 à 1990. C'est donc une ville assez ancienne.

%matplotlib inline

Nb_ResPApres2011_LaChapelleSaintMaurice = tab.loc[tab['Nom commune']=="La Chapelle-Saint-Maurice"]

df_R1919 = Nb_ResPApres2011_LaChapelleSaintMaurice[["Nom commune","Résidences principales avant 1919"]].groupby(['Nom commune']).mean()
df_R19191945 = Nb_ResPApres2011_LaChapelleSaintMaurice[["Nom commune","Résidences principales de 1919 à 1945"]].groupby(['Nom commune']).mean()
df_R19461970 = Nb_ResPApres2011_LaChapelleSaintMaurice[["Nom commune","Résidences principales de 1946 à 1970"]].groupby(['Nom commune']).mean()
df_R19711990 = Nb_ResPApres2011_LaChapelleSaintMaurice[["Nom commune","Résidences principales de 1971 à 1990"]].groupby(['Nom commune']).mean()
df_R19912005 = Nb_ResPApres2011_LaChapelleSaintMaurice[["Nom commune","Résidences principales de 1991 à 2005"]].groupby(['Nom commune']).mean()
df_R20062010 = Nb_ResPApres2011_LaChapelleSaintMaurice[["Nom commune","Résidences principales de 2006 à 2010"]].groupby(['Nom commune']).mean()
df_R2011 = Nb_ResPApres2011_LaChapelleSaintMaurice[["Nom commune","Résidences principales après 2011"]].groupby(['Nom commune']).mean()


display(df_R1919)
display(df_R19191945)
display(df_R19461970)
display(df_R19711990)
display(df_R19912005)
display(df_R20062010)
display(df_R2011)
```
