```python
#Calcul du classement des 20 meilleurs villes avec la plus forte évolution de la consommation électrique résidentielle et modélisation à l'aide d'un histogramme

import pandas as pd
%matplotlib inline

communes = pd.read_csv("filtered-data.csv", delimiter = ";")


conso_evolution_resi = communes[['Commune','Diff_ConsoResidentiel']]
conso_evolution_resi.groupby("Commune").agg({"Diff_ConsoResidentiel":"sum"}).Diff_ConsoResidentiel.nlargest(20).plot(kind="bar", title="Classement des 20 meilleurs villes avec la plus forte évolution de la consommation électrique résidentielle", legend=True)
```

```python
#Calcul du classement des 20 meilleurs villes avec la plus forte évolution de la consommation électrique industrielle et modélisation à l'aide d'un histogramme

conso_evolution_indus = communes[['Commune','Diff_ConsoIndustriel']]
conso_evolution_indus.groupby("Commune").agg({"Diff_ConsoIndustriel":"sum"}).Diff_ConsoIndustriel.nlargest(20).plot(kind="bar", legend=True, title="Classement des 20 meilleurs villes avec la plus forte évolution de la consommation électrique industrielle")
```

```python
#Calcul du classement des 20 meilleurs villes avec la plus forte évolution de la consommation électrique tertiaire et modélisation à l'aide d'un histogramme

conso_evolution_tertiaire = communes[['Commune','Diff_ConsoTertiaire']]
conso_evolution_tertiaire.groupby("Commune").agg({"Diff_ConsoTertiaire":"sum"}).Diff_ConsoTertiaire.nlargest(20).plot(kind="bar", legend=True, title="Classement des 20 meilleurs villes avec la plus forte évolution de la consommation électrique tertiaire")
```
